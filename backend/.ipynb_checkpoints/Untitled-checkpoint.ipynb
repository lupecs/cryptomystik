```python
#

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Fetch Bitcoin data
btc = yf.Ticker("BTC-USD")
btc_data = btc.history(period="1y")

# Plot Closing Prices
plt.figure(figsize=(12,5))
plt.plot(btc_data.index, btc_data["Close"], label="BTC Closing Price")
plt.legend()
plt.show()

# Display Data
btc_data.head()
```
